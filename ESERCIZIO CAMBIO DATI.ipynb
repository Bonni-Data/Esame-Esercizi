```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
ys= 200*10 + np.random.randn(10)
x= [x for x in range(len(ys))]
```

```python
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 1995, where=(ys>195), facecolor='b', alpha= 0.5)
plt.title("Ma saranno cazzi miei?")
plt.show()
```
